```r
library(dplyr)
library('fixest')
options(warn=-1)
library(modelsummary)
library(estimatr)
library(tidyverse)
library(rmarkdown)
```

```r
data = read.csv('data_20220301/data_did_by_contest.csv')
```

```r
head(data)
```

```r
names(data)
```

```r
regression <- list()

regression[[4]] <- fixest::feols(number_of_participants ~ treated + after + treated * after + prize 
                                 + guaranteed + blind + multi_winner,
                                 vcov = "hetero", data = data)

regression[[5]] <- fixest::feols(total_submission ~ treated + after + treated * after + prize 
                                 + guaranteed + blind + multi_winner, 
                                 vcov = "hetero", data = data)

regression[[6]] <- fixest::feols(total_submission ~ treated + after + treated * after + prize 
                                 + guaranteed + blind + multi_winner + number_of_participants, 
                                 vcov = "hetero", data = data)

regression[[7]] <- fixest::feols(number_of_submission_per_participants ~ treated + after + treated * after + prize 
                                 + guaranteed + blind + multi_winner, 
                                 vcov = "hetero", data = data)

esttable(regression, order = c("winning_rate_feb_perc", "^treated x after$", "!(Intercept)"),se = "hetero", digits = "r2")
```
