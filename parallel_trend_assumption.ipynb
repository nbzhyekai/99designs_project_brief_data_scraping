```r
library("xlsx")
library(dplyr)
library('fixest')
options(warn=-1)
library(modelsummary)
library(estimatr)
library(tidyverse)
library(rmarkdown)
```

```r
# data <- xlsx::read.xlsx("data_20220301/data_did_by_contest.csv",     # Read xlsx file with read.xlsx
#                    sheetIndex = 1)
data = read.csv("data_20220301/parallel.csv")
```

```r
head(data)
```

```r
names(data)
```

```r
regression <- list()

regression[[1]] <- fixest::feols(number_of_participants ~ treated + I(week_pre_6 * treated) 
                                 + I(week_pre_5 * treated) 
                                 + I(week_pre_4 * treated) + I(week_pre_3 * treated)
                                 + I(week_pre_2 * treated) +  I(week_post_1 * treated) 
                                 + I(week_post_2 * treated) + I(week_post_3 * treated) + I(week_post_4 * treated)
                                 + I(week_post_5 * treated) + I(week_post_6 * treated), 
                                 vcov = "hetero", data = data)

regression[[2]] <- fixest::feols(total_submission ~  treated + I(week_pre_6 * treated) 
                                 + I(week_pre_5 * treated) 
                                 + I(week_pre_4 * treated) + I(week_pre_3 * treated)
                                 + I(week_pre_2 * treated) +  I(week_post_1 * treated) 
                                 + I(week_post_2 * treated) + I(week_post_3 * treated) + I(week_post_4 * treated)
                                 + I(week_post_5 * treated) + I(week_post_6 * treated), 
                                 vcov = "hetero", data = data)

regression[[3]] <- fixest::feols(total_submission ~  treated +  treated + I(week_pre_6 * treated) 
                                 + I(week_pre_5 * treated) 
                                 + I(week_pre_4 * treated) + I(week_pre_3 * treated)
                                 + I(week_pre_2 * treated) +  I(week_post_1 * treated) 
                                 + I(week_post_2 * treated) + I(week_post_3 * treated) + I(week_post_4 * treated)
                                 + I(week_post_5 * treated) + I(week_post_6 * treated) 
                                 + number_of_participants, 
                                 vcov = "hetero", data = data)
regression[[4]] <- fixest::feols(number_of_submission_per_participants ~  treated +  treated + I(week_pre_6 * treated) 
                                 + I(week_pre_5 * treated) 
                                 + I(week_pre_4 * treated) + I(week_pre_3 * treated)
                                 + I(week_pre_2 * treated) +  I(week_post_1 * treated) 
                                 + I(week_post_2 * treated) + I(week_post_3 * treated) + I(week_post_4 * treated)
                                 + I(week_post_5 * treated) + I(week_post_6 * treated) , 
                                 vcov = "hetero", data = data)

esttable(regression, order = c("winning_rate_feb_perc", "^treated x after$", "!(Intercept)"),se = "hetero", digits = "r2")
```
